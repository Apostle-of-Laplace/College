```python
# Importing Libraries

import os
import warnings 
import matplotlib.pyplot as plt
import pandas as pd

warnings.filterwarnings('ignore')
```

```python
# Importing and Reading CSV Data Functions
 
def file_extractor(file_name):

    file_path = os.path.join(os.getcwd(), file_name)
    
    with open(file_path, 'r') as f:
        lines = f.readlines()

    data_lines = [line for line in lines if len(line.strip().split()) >= 2 and line.strip()[0].isdigit()]
    
    from io import StringIO

    data_str = ''.join(data_lines)

    df = pd.read_csv(StringIO(data_str), delim_whitespace=True, header=None)
    
    df = df.iloc[:, :2]

    df.columns = ['Raw Signal', 'Filtered Signal']

    return df
```

```python
# Plotting Data Functions

def plot_signals(df, title):
    
    plt.figure(figsize=(10, 5))
    
    plt.plot(df['Raw Signal'], label='Raw Signal')
    
    plt.plot(df['Filtered Signal'], label='Filtered Signal')
    
    plt.xlabel('Sample')
    
    plt.ylabel('Signal Value')
    
    plt.title(title)
    
    plt.legend()
    
    plt.tight_layout()
    
    plt.show()
```

```python
# Running the Script

csv_files = [ 'ws3.csv', 'ws6.csv', 'ws9.csv', 'ws20.csv' ]

for file_name in csv_files:
    
    df = file_extractor(file_name)

    plot_signals(df, title=f"Signals from {file_name}")
```
